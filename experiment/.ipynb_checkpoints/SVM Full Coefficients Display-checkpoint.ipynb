```python
# check feature importance

def f_importances(coef, names):
    imp = coef
    imp, names = zip(*sorted(zip(imp, names)))
    plt.barh(range(20), imp[:10] + imp[-10:], align='center')
    plt.yticks(range(20), names[:10] + names[-10:])
    plt.show()
```

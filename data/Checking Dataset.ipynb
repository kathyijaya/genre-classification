```python
dataset_wkspc = os.path.join(ai_lit_path, 'workspace', 'gb_input')
train_json_path = os.path.join(dataset_wkspc, 'train_index_list.json')
test_json_path = os.path.join(dataset_wkspc, 'test_index_list.json')

with open(train_json_path) as f:
    train_json = json.load(f)
with open(test_json_path) as f:
    test_json = json.load(f)
```

```python
train_subjects = {}
for entry in train_json:
    train_subjects.update({entry['index']: entry['subjects']})
```

```python
train_subjects
```

```python
subjects = gb_input.get_subjects(dataset_wkspc)

for key in train_subjects:
    train_subjects.update({key: list(set(train_subjects[key]).intersection(subjects))})
```

```python
more_than_one_genre = []
for key, value in train_subjects.items():
    if (len(value) > 1):
        more_than_one_genre.append(key)
```
